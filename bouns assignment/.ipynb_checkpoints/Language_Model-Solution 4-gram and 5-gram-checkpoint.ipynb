```python
import re
f = open("9053.txt", "r")
Dataset = []
for line in f:
   line = re.sub(r'\@\@\d{8} \@\d{7}/','',line)
   line = re.sub(r'<h>|<p>','',line)
   line = re.sub(r'  ',' ',line)
   line = re.sub(r'\@','',line)
   Dataset.append(line)
```

```python
def Bigram(string):
    string = "<START> "+string+" <END>"
    word_list = string.split(' ')
    bigrams = []
    for i in range(len(word_list)-1):
        bigrams.append([word_list[i],word_list[i+1]])
    return bigrams
```

```python
string = "Hello world"
assert len(Bigram(string)) == 3
```

```python
def Trigram(string):
    string = "<START> "+ string +" <END>"
    word_list = string.split(' ')
    Trigrams = []
    for i in range(len(word_list)-2):
        Trigrams.append([word_list[i],word_list[i+1],word_list[i+2]])
    return Trigrams
```

```python
string = "Hello world today is good"
assert len(Trigram(string)) == 5
print(Trigram(string))
```

```python
from collections import Counter, defaultdict
Total_Count = 0
Trigram_Dict = defaultdict(lambda: defaultdict(lambda: 0))
for line in Dataset:
    list_of_grams = Trigram(line)
    Total_Count += len(list_of_grams)
    for w1,w2,w3 in list_of_grams:
      Trigram_Dict[(w1, w2)][w3] += 1
```

```python
Trigram_Dict
```

```python
for w1_w2 in Trigram_Dict:
  for w3 in Trigram_Dict[w1_w2]:
      Trigram_Dict[w1_w2][w3] /=  Total_Count
```

## **4-gram**

```python
def fourgram(string):
    string = "<START> "+string+" <END>"
    word_list = string.split(' ')
    fourgram = []
    for i in range(len(word_list)-3):
        fourgram.append([word_list[i],word_list[i+1],word_list[i+2] , word_list[i+3] ] )
    return fourgram
```

```python
string = "Hello world today is good"
print(fourgram(string))
```

```python
from collections import Counter, defaultdict
Total_Count = 0
Fourgram_Dict = defaultdict(lambda: defaultdict(lambda: 0))
for line in Dataset:
    list_of_grams = fourgram(line)
    Total_Count += len(list_of_grams)
    for w1,w2,w3,w4 in list_of_grams:
       Fourgram_Dict[(w1, w2 , w3)][w4] += 1
```

```python
for w1_w2_w3 in Fourgram_Dict:
    for w4 in Fourgram_Dict[w1_w2_w3]:
      Fourgram_Dict[w1_w2_w3][w4] /=  Total_Count
```

## **5-gram**

```python
def fivegram(string):
    string = "<START> "+string+" <END>"
    word_list = string.split(' ')
    fivegram = []
    for i in range(len(word_list)-4):
        fivegram.append([word_list[i],word_list[i+1],word_list[i+2] , word_list[i+3] ,word_list[i+4]  ] )
    return fivegram
```

```python
string = "Hello world today is good"
print(fivegram(string))
```

```python
from collections import Counter, defaultdict
Total_Count = 0
Fivegram_Dict = defaultdict(lambda: defaultdict(lambda: 0))
for line in Dataset:
    list_of_grams = fivegram(line)
    Total_Count += len(list_of_grams)
    for w1,w2,w3,w4,w5 in list_of_grams:
        Fivegram_Dict[(w1, w2 , w3,w4)][w5] += 1
```

```python
for w1_w2_w3_w4 in Fivegram_Dict:
    for w5 in Fivegram_Dict[w1_w2_w3_w4]:
        Fivegram_Dict[w1_w2_w3_w4][w5] /=  Total_Count
```

```python
s = ["<START>","","Welcome","to"]

for i in range(25):
    next_word = random.choices(list(Fivegram_Dict[s[-4],s[-3],s[-2],s[-1]].keys()))
    s.append(next_word[0])
print(s)
```

**Comparing between 4-gram and 5-gram**

Answer : The sentences in 4-gram is much more realistic and make more sence. Moreover , it is generic as it at every time you run the code it gives you a new sentence and it also make sence most of the time. However , for the 5-gram the sentence is fixed all the time as it give the same sentence every time you run the code. In addition to that the sentence doesn't make sence.
